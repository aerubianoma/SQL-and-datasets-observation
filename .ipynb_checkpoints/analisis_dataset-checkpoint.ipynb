Para esta prueba escogi el dataset de Students Performance, en la sgte celda podemos observar como se ve el dataset en una primera instancia.

```python
# Libreria que nos permite manipular csv comodamente en python
import pandas as pd
# Leemos el csv con ayuda de la libreria pandas
df = pd.read_csv ('StudentsPerformance.csv')
display(df)
```

En este dataset podemos encontrar la calificacion de estudiantes cada uno de estos clasificado por:

- Genero
- Raza/Etnia
- Nivel de educacion de los padres
- Tipo de almuerzo que reciben
- Si tomaron o no preparacion para el test
- Calificacion en cada una de las areas del test (matematicas,lectura,escritura)

A continuacion se añadira una nueva columna en la cual se encuentra la calificacion en promedio de cada estudiante

```python
# añadimos la nueva columna con el promedio de las calificaciones
df = df.assign(average = (df['math score']+df['reading score']+df['writing score'])/3)
display(df)
```

# Analisis del dataset

Observemos como afecta cada caracteristica a la calificacion Promedio de los estudiantes y con esto obtengamos conclusiones

## Grupo Etnico

```python
# Libreria util para realizar graficos
import matplotlib.pyplot as plt

# Grupos etnicos en el dataset
etnias = df['race/ethnicity'].unique()
averages = []

# Obtenemos el promedio por grupo etnico
for etnia in etnias:
    averages.append(df[df['race/ethnicity']==etnia]['average'].mean())

# Para quitar los bordes de la grafica
'''ax = plt.axes()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)'''

# Creamos Gráfica
plt.barh(etnias, averages, color = 'blue')
 
# Título de Gráfica
plt.title('Calificacion en promedio por etnia')
 
# Por si se quiere ver el valor exacto en cada grupo etnico
'''for index, value in enumerate(averages): 
    plt.text(value, index, 
             str(value)) '''
    
## Mostramos Gráfica
plt.show()
```

## Genero

```python
# Libreria util para realizar graficos
import matplotlib.pyplot as plt

# Grupos etnicos en el dataset
genders = df['gender'].unique()
averages = []

# Obtenemos el promedio por genero
for gender in genders:
    averages.append(df[df['gender']==gender]['average'].mean())

# Para quitar los bordes de la grafica
'''ax = plt.axes()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)'''

# Creamos Gráfica
plt.barh(genders, averages, color = 'green')
 
# Título de Gráfica
plt.title('Calificacion en promedio por genero')
 
# Por si se quiere ver el valor exacto en cada grupo etnico
'''for index, value in enumerate(averages): 
    plt.text(value, index, 
             str(value)) '''
    
## Mostramos Gráfica
plt.show()
```

# Primeras conclusiones

De los analisis anteriores podemos observar que:

- El realizar un test de preparacion mejora el rendimiento de los estudiantes
- Los estudiantes que reciben un almuerzo estandar tienen mejores calificaciones
- El nivel de estudio de los padres afecta significativamente el rendimiento de los estudiantes, a mas nivel, mejor calificacion
- Las chicas obtuvieron mejores calificaciones, sin embargo es una metrica muy igualada
- Los estudiantes del grupo E obtuvieron mejores calificaciones
